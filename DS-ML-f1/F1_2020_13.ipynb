```python
df_ThirteenP1_2020 = pd.read_html(ThirteenP1_2020, header=0)
df_ThirteenQ_2020 = pd.read_html(ThirteenQ_2020, header=0)
df_ThirteenRR_2020 = pd.read_html(ThirteenRR_2020, header=0)

df_ThirteenP1_2020 = df_ThirteenP1_2020[0]
df_ThirteenQ_2020 = df_ThirteenQ_2020[0]
df_ThirteenRR_2020 = df_ThirteenRR_2020[0]
```

```python
def uniquify(change_dups):
    seen = set()
    
    for dup in change_dups:
        name = 1
        new_name = dup
        
        while new_name in seen:
            name += 1
            new_name = "{}_{}".format(dup, name)
            
        yield new_name
        seen.add(new_name)
        
list(uniquify(change_dups))
```
