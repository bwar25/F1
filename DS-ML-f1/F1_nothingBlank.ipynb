```python
def change_values(race):
    for col, row in race.iteritems():
        if isinstance(row[0], str):
            race.loc[race[col] == 'n/a', col] = 0
            race.loc[race[col] == 'NC', col] = 0
            race.loc[race[col] == 'DNF', col] = 0
            
def no_NA(race):
    race = race.replace(to_replace='n/a',
                value='0', inplace=True)
    
def no_NC(race):
    race = race.replace(to_replace='NC',
                value='0', inplace=True)
    
def no_DNF(race):
    race = race.replace(to_replace='DNF',
                value='0', inplace=True)
    
def no_DNS(race):
    race = race.replace(to_replace='DNS',
                value='0', inplace=True)
    
def no_lap(race):
    race = race.replace(to_replace='+1 lap', 
                value='0', inplace=True)
    
def no_laps(race):
    race = race.replace(to_replace='+2 laps', 
                value='0', inplace=True)
```

```python
for race in all_races:
    no_NA(race)
    no_NC(race)
    no_DNF(race)
    no_DNS(race)
    no_lap(race)
    no_laps(race)
```
